```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pd_df = pd.read_csv('company_sales_data.csv')
pd_df.head()
```

```python
profit_list = pd_df['total_profit'].tolist()
profit_list
```

```python
month_list = pd_df['month_number']
month_list
```

```python
units_list = pd_df['total_units'].tolist()
units_list
```

```python
plt.plot(month_list, profit_list)
plt.title('распределение прибыли по месяцам')
plt.xlabel('месяцы')
plt.ylabel('прибыль')
plt.xticks(month_list)
plt.yticks([100000, 300000, 500000])
plt.show
```

```python
plt.plot(month_list, units_list, label = 'информация о продажах', color = 'red', linestyle = '--', marker = 'o', 
         markerfacecolor = 'k', linewidth = 3)
plt.title('кол-во проданных единиц')
plt.xlabel('месяцы')
plt.ylabel('кол-во')
plt.xticks(month_list)
#plt.yticks([100000, 300000, 500000])
plt.legend(loc = 'lower right')
plt.show
```

```python
toothpaste_list = pd_df['toothpaste'].tolist()
toothpaste_list
```

```python
plt.scatter(month_list, toothpaste_list, label = 'кол-во штук')
plt.title('кол-во проданных единиц tothpaste')
plt.xlabel('месяцы')
plt.ylabel('кол-во')
plt.xticks(month_list)
plt.grid(linestyle = '--')
#plt.yticks([100000, 300000, 500000])
plt.legend(loc = 'lower right')
plt.show
```

```python
facecream_list = pd_df['facecream'].tolist() #tolist - принудительно превращает в список что бы не было других толкований
facecream_list
```

```python
facewash_list = pd_df['facewash'].tolist()
facewash_list
```

```python
# plt.bar([value - 0.2 for value in month_list], facecream_list, color = 'r')
# plt.bar([value + 0.2 for value in month_list], facewash_list, color = 'b')
plt.bar(month_list - 0.2, facecream_list, color = 'r', width = 0.5, label ='Средство для лица')
plt.bar(month_list + 0.2, facewash_list, color = 'b', width = 0.5, label = 'Средство для умывания')
plt.xticks(month_list)
plt.legend()
plt.grid(linestyle = '--')
plt.show
```

```python
bathingsoap_list = pd_df['bathingsoap'].tolist()
bathingsoap_list
```

```python
shampoo_list = pd_df['shampoo'].tolist()
shampoo_list
```

```python
moisturizer_list = pd_df['moisturizer'].tolist() #увлажняющий крем
moisturizer_list
```

```python
# №1
plt.plot(month_list, facecream_list, color = 'r', linestyle = '--', marker = 'o', 
         markerfacecolor = 'w', linewidth = 2, label = 'крем для лица')
plt.plot(month_list, facewash_list, color = 'b', linestyle = '-', marker = 's', 
         markerfacecolor = 'b', linewidth = 2, label = 'с-во для умывания')
plt.plot(month_list, toothpaste_list, color = 'g', linestyle = '--', marker = 'o', 
         markerfacecolor = 'w', linewidth = 2, label = 'зубная паста')
plt.plot(month_list, bathingsoap_list, color = 'c', linestyle = '--', marker = 'o', 
         markerfacecolor = 'w', linewidth = 2, label = 'мыло')
plt.plot(month_list, shampoo_list, color = 'm', linestyle = '--', marker = 'o', 
         markerfacecolor = 'w', linewidth = 2, label = 'шампунь')
plt.plot(month_list, moisturizer_list, color = 'y', linestyle = '--', marker = '.', 
         markerfacecolor = 'w', linewidth = 2, label = 'увлажняющий крем')
plt.title('данные о продажах продуктов')
plt.xlabel('месяцы')
plt.ylabel('кол-во шт.')
plt.xticks(month_list)
plt.yticks([i for i in range(1000,16000,1000)])
plt.legend(loc='upper left', bbox_to_anchor=(1.0, 1.0))
plt.show
# Примечание: графики "с-во для умывания" и "увлажняющий крем" совпадают
```

```python
# №2
plt.bar(month_list, bathingsoap_list, color = 'c', width = 0.5, label ='мыло')
plt.xticks(month_list)
plt.legend()
```

```python
# №3
fcr_total = sum(facecream_list)
fcw_total = sum(facewash_list)
tpaste_total = sum(toothpaste_list)
soap_total = sum(bathingsoap_list)
shampoo_total = sum(shampoo_list)
mst_total = sum(moisturizer_list)
```

```python
values = [fcr_total, fcw_total, tpaste_total, soap_total, shampoo_total, mst_total]
labels = ['крем для лица', 'с-во для умывания', 'зубная паста', 'мыло', 'шампунь', 'увлажняющий крем']
colors = ['r', 'b', 'g', 'c', 'm', 'y']
plt.pie(values, labels = labels, colors = colors, autopct='%1.2f%%')
plt.legend(loc='upper left', bbox_to_anchor=(1.5, 1.0))

plt.show
```
